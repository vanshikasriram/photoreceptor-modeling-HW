```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = .01
t = np.arange(0, 6+dt, dt)

def h1(t):
    return t / 0.2 * np.exp(-t / 0.2)

def light_intensity(a):
    z = np.zeros_like(t)
    n = int(t[-1]/2)
    
    for i in range(n+1):
        for j in range(len(t)):
            if (2*(i-1) <= t[j] <= 2*i):
                z[j] = a*(2**(i))
    return z
```
